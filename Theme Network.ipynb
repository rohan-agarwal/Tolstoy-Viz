```python
import csv 
import networkx as nx
import matplotlib.pyplot as plt 
import itertools as it

%matplotlib inline
```

```python
f = "C:/Users/Rohan/Documents/GitHub/Tolstoy-Viz/themes_quotes.csv"

nodelist = range(1,26)
edgelist = {}

with open(f, 'rb') as cf: 
    r = csv.reader(cf, delimiter = ",")
    for row in r:
        edges = row[2].split(', ')
        edges = [int(a) for a in edges]
        if len(edges) > 1:
            edgelist[row[0]] = list(it.combinations(edges, 2))
```

```python
print nodelist
print len(nodelist)
print edgelist
print len(edgelist)
```

```python
G = nx.MultiGraph()
G.add_nodes_from(nodelist)
edgeweights = {}
for key, edges in edgelist.iteritems():
    for edge in edges:
        G.add_edge(edge[0], edge[1], color=int(key))
```

```python
print G.number_of_nodes()
print G.number_of_edges()
```

```python
deg = nx.degree(G)
nodesizes = [max(400, v * 100) for v in deg.values()]

edges,weights = zip(*nx.get_edge_attributes(G,'color').items())
```

```python
graph_pos = nx.shell_layout(G)
node_alpha = 1
node_color = '#6699cc'
node_size = nodesizes
edge_thickness = 1.5
edge_alpha = 0.9
node_text_size = 20
text_font = "arial"
text_color = '#fff7e7'

plt.figure(figsize=(20,20))
plt.axis('off')

nx.draw_networkx_nodes(G, graph_pos,node_size=node_size, 
                       alpha=node_alpha, node_color=node_color,
                       nodelist = deg.keys())
nx.draw_networkx_edges(G,graph_pos,width=edge_thickness,
                       alpha=edge_alpha,edgelist=edges,
                       edge_color=weights,edge_cmap=plt.cm.winter)
nx.draw_networkx_labels(G, graph_pos,font_size=node_text_size,
                        font_family=text_font, font_color = text_color)

plt.savefig("theme_network.png",bbox_inches='tight')
```
